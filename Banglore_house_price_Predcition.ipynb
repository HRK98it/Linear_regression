# Business problem

**Predict the Price of Bangalore House**

Using linear Regression- Supervised Machine learning Algorithm

```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB"
df=pd.read_csv(path)
```

```python
pd.set_option('display.max_column',None)
pd.set_option('display.max_row',None)

df.head()
```

Spliting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('price',axis=1)
y=df['price']
print("Shape of X ",X.shape)
print('Shape of y: ',y.shape)
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=51)
print("Shape of X_train: ",X_train.shape)
print("Shape of X_test: ",X_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test: ",y_test.shape)
```

Now we will do feature Scaling as in our dataset there are many features which has numeric value not in specific range

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_train)

X_train = sc.transform(X_train)
X_test= sc.transform(X_test)
```

Training the machine learnig model
Linear Regression ML model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train) #it will learn the value of intercept and coefficient like (B1,B2 etc)
```

```python
# checking which value model has learn

lr.coef_  #it will show the vlaue of (B!,B2 etc)
```

```python
# checking which intercept value the model has learn

lr.intercept_
```

#predict the value of home and test

```python
# testing fromm the test data
# 1st house imformation with all column

X_test[0,:]

lr.predict([X_test[0, :]])  #it is showing the value 76 means it predict the price of house that is 76 lakh
```

```python
# predicting the price of all house it the X_test data

lr.predict(X_test)
```

```python
# checking the accuracy of model

lr.score(X_test, y_test) # the accuracy is 79% which is best for the regression model
# we can fine tuned it also (using hyper parameter or different model)
```
